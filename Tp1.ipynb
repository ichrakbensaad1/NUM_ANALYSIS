```python
def dicho(a,b,f,n):
    '''version itérative '''
    m = (a+b)/2
    err = abs(b-a)
    while (err>n):
        if f(m)==0:
          break
        if f(a) * f(m) <= 0: 
            b=m
        else:
            a=m
        m=(a+b)/2
        err=abs(b-a)
    
    return m


  

```

version tant que qui renvoila solution avec le nombre d'''itération


```python
def dichow(a,b,f,eps):  
    """version tant que qui renvoi la solution avec le nombre d'''itération """
    c , d = min (a , b ) , max (a , b )
    valc , vald = f ( c ) , f ( d )
    n = 0
    while d - c > eps :
        n += 1
        m = ( c + d ) / 2
        valm = f ( m )
        if valc * valm < 0 :
            d = m
            vald = valm
        else :
            c = m
            valc = valm
    print (" Le nombre d'itérations est :", n )
    print ("Valeur approchée est",( c + d ) / 2)
    return n
  
```

```python
def ValeurApproch(a,b,f,eps,n):
    x=a+b/2
    while(n>0):
        if(f(a)*f(b)<0):
            b=x
        if(f(b)*f(x)<0):
            a=b
            b=x
        n = n-1
    return (a+b)/2
```

```python
#𝛼1∈]−2;−1.5[ 
a=-2
b=-1.5
eps=0.0001
n = dichow(a,b,f,eps)
print(ValeurApproch(a,b,f,eps,n))
#𝛼2∈]−1.5;−0.5[
a=-1.5
b=-0.5
eps=0.0001
n = dichow(a,b,f,eps)
print(ValeurApproch(a,b,f,eps,n))
#𝛼3∈]1.5;2[
a=1.5
b=2
eps=0.0001
n = dichow(a,b,f,eps)
print(ValeurApproch(a,b,f,eps,n))
```

1) On considère la fonction $f(x)=e^{-x}-x$. Ecrire puis appliquer une fonction Matlab que l'on appellera newton.m à la >fonction $f$, en prenant $x0=0$, $eps=0.001$ .

2) Montrer que l'équation $f(x)=0$ admet une solution unique que l'on peut obtenir en utilisant la méthode de Newton à partir >de $x0=1$.

**1)**

```python
from math import exp
import matplotlib.pyplot as plt
from numpy import linspace 
import numpy as np
```

```python
def newton(a,eps):
    x1=a
    x2=x1-f(x1)/fprime(x1)
    while abs(x1-x2) > eps:
        x1=x2
        x2=x1-f(x1)/fprime(x1)
    return x2
```

```python
f= lambda x: np.exp(-x) - x
fprime= lambda x: -np.exp (x) - 1 
T=linspace(0,1,10)
plt.title("solution dans [-2:2] ")
plt.xlabel("x")
plt.ylabel ("f(x)")
plt.grid(True)
plt.plot(T,f(T))
```
